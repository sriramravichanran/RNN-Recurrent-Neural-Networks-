34) TRANSFORMER ARCHITECTURE
· Embedding: Converts input tokens into dense vector representations to capture semantic meaning.
· Positional Encoding: Adds positional information to embeddings since the model has no recurrence, using sine/cosine functions.
· Multi-Head Attention: Runs multiple self-attention heads in parallel to capture different relationships, computed as
    Attention(Q,K,V) = softmax(QK^T/√d_k)V.
· Feed Forward Network: A two-layer fully connected network with ReLU activation applied independently at each position.
· Residual Connections: Skip connections around each sub-layer to enable better gradient flow.
· Layer Normalization: Normalizes activations across features for stable and faster training.
